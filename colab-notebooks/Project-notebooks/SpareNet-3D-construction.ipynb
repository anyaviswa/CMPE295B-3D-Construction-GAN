## Model Inferencing

### How to change input data path?
test.py takes all its arguments from configs/base_config.py. In this file, the following fields can be manipulated to modify train and test data.

<b> Training data </b> <br>
__C.DATASETS.shapenet.partial_points_path <br>
__C.DATASETS.shapenet.complete_points_path <br>
__C.DATASETS.shapenet.bounding_box_file_path <br>

<b> Test data </b> <br>
__C.DATASETS.kitti.category_file_path <br>
__C.DATASETS.kitti.partial_points_path <br>
__C.DATASETS.kitti.bounding_box_file_path <br>

SpareNet model expects boundary boxes for every image both during training and testing. Not all images had boundary box by default. For this project, boundary boxes are identified for those images which lacks it. This boundary boxes calculation can be seen in colab-notebooks/helper-notebooks/3D_construction_helper.ipynb file.

### How to manipulate model parameter?
configs/ folder encloses yaml file for each model which consists of all model parameters.
