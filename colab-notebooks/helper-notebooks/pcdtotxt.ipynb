```python
import os
import numpy as np
import open3d as o3d
dir_name='/content/drive/MyDrive/ShapeNetCompletion/ShapeNetCompletion/train/complete/02958343'
counter=0
for file in os.listdir(dir_name):
  # print(file)
  file_name = dir_name+'/'+file
  counter+=1
  pcd= o3d.io.read_point_cloud(file_name)
  array=np.asarray(pcd.points)
  # print(array)
  with open("/content/drive/MyDrive/ShapeNetCompletionText1/car_"+str(counter)+".txt", mode='w') as f:  
    for i in range(len(array)):
        f.write("%f,"%float(array[i][0].item()))
        f.write("%f,"%float(array[i][1].item()))
        f.write("%f,"%float(array[i][2].item()))
        f.write("%f,"%float(array[i][0].item()))
        f.write("%f,"%float(array[i][1].item()))
        f.write("%f\n"%float(array[i][2].item()))
```
